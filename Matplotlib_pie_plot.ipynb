
# 2 Pie plot


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))

#colors=["lightskyblue","steelblue","lightblue","sandybrown","mediumseagreen", "rosybrown", "darkgrey","coral","pink",  "plum","lightsalmon","salmon","mediumaquamarine"]
plt.rcParams.update({'font.size': 16})

data=[10,20,30,40] # 数据
labels=["A","B","C","D"]
colors=["lightskyblue","coral", "mediumseagreen","lightblue"]
plt.pie(data,labels=labels,colors=colors,startangle=0, shadow=True,explode=(0,0.1,0,0.05),  autopct='%0.1f%%')   
plt.title('Pie Plot')
plt.show()
```
