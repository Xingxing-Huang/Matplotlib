
# 1 Horizontal bar chart


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcdefaults()
fig, ax = plt.subplots()

code = ('C', 'C++', 'Python', 'Java', 'JS')
y_pos = np.arange(len(code))
people = ([ 8.3,  4.9, 11.6 , 12.7, 11.8])
 
ax.barh(y_pos, people, color='', edgecolor='black', align='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(code)
ax.invert_yaxis()  # labels read top-to-bottom

ax.set_xlabel('People (million)')
ax.set_ylabel('Code')
ax.set_title('How many people use them in 2050?')

plt.show()
```
